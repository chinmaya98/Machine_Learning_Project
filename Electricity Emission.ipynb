```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Electricity original.csv')

# Carbon intensity for California (gCO2e/MJ) for the year 2022
carbon_intensity = 76.73  # gCO2e/MJ

# Convert carbon intensity from gCO2e/MJ to gCO2/kWh (assuming 1 MJ = 0.277778 kWh)
# Function to calculate carbon emissions
def calculate_emission(row):
    total_consumption = row['Monthly Consumption (kWh)'] * 3.6  # Convert kWh to MJ
    emission = total_consumption * carbon_intensity / 1000  # Convert to grams
    return round(emission, 2)

# Create a new column 'Emission_Electricity' and calculate emissions for each row
df['Emission_Electricity'] = df.apply(calculate_emission, axis=1)

# Save the updated DataFrame back to the CSV file
df.to_csv('Electricity.csv', index=False)

print("Emission values added to the CSV file.")
```
